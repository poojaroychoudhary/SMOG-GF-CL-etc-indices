**FUNCTION TO COUNT SYLLABLES IN EACH WORD PROBLEM**

**Counted number of syllables.**

```python
def syllable_count(df_col):
    df_col = df_col.lower()
    count = 0
    vowels = "aeiou"
    if df_col[0] in vowels:
        count += 1
    for index in range(1, len(df_col)):
        if df_col[index] in vowels and df_col[index - 1] not in vowels:
            count += 1
    if df_col.endswith("e"):
        count -= 1
    if count == 0:
        count += 1
    return count
```

```python
df_data['syllables'] = df_data['Problem'].apply(syllable_count)
```

```python
df_data.head()
```

**FUNCTION TO COUNT OF POLY SYLLABLES IN EACH WORD PROBLEM**

```python
def count_poly_syllables(df_col):
  total_count = 0
  df_col = df_col.split()
  for word in df_col: 
    #print(word)
    temp = syllable_count(word)
    if temp>=3:
      total_count = total_count+temp
  return total_count
```

```python
df_data['poly_syllables'] = df_data['Problem'].apply(count_poly_syllables)
```

**CALCULATION OF DIFFERENT INDICES**
1. Gunning Fog Index
2. Flesch - Kincaid Readability Index
3. SMOG (Simple Measure of Gobbledygook) Index
4. Automated Readability Index
5. Coleman-Liau Index

**Gunning Fog Index**
0.4[ (words/sentence) + 100(TMVC/words) ]

```python
# 0.4[ (words/sentence) + 100(TMVC/total words) ] 
#(words/sentence) = ASL
#(TMVC/total words) = PMW

GF = []
for i in df.iloc[:,0]:
  w = count_allwords(i)
  c = word_length(i)
  s = count_sentences(i)
  gf_score = (w/s) + (100*c/w)
  GF.append(0.4*gf_score)
```

```python
len(GF)
```

**Flesch - Kincaid Readability Index**

```python
#Formula: 206.835 - 1.015(total words/total sentences) - 84.6(TMVC/total words)
#(total words/total sentences) = ASL
FK = []
for i in df.iloc[:,0]:
  w = count_allwords(i)
  c = word_length(i)
  s = count_sentences(i)
  fk_score = (1.015*w/s) + (84.6*c/w)
  fk_score = 206.835 - fk_score 
  FK.append(fk_score) 
```

```python
len(FK)
```

**SMOG (Simple Measure of Gobbledygook) Index**

```python
# Fromula: 3.1291 + 1.0430(sqrt( TMVC*30 /number of sentences/number of sentences))
SMOG = []
for i in df.iloc[:,0]:
  c = word_length(i)
  s = count_sentences(i)
  smog_score = np.sqrt(30*c/s/s)
  smog_score = smog_score * 1.0430
  SMOG.append(3.1291+smog_score) 
```

```python
len(SMOG)
```

**Automated Readability Index**

4.71 x (CMVW/TMVC) + 0.5 x (TMVC/ no of sentences) – 21.43.

```python
AR = []
for i in df.iloc[:,0]:
  c = word_length(i)
  w = count_words(i)
  s = count_sentences(i)
  ar_score = (4.71*w/c) + (0.5*c/s) - 21.43
  AR.append(ar_score) 
```

```python
len(AR)
```

**Coleman-Liau Index**

```python
#Formula: 0.0588(average number of math vocab words per 25 words) - 15.8 - 0.296(average number of sentences per 25 words)
#Formula: 0.0588(AMV) - 15.8 - 0.296(AMVS)
CL = []
for i in df.iloc[:,0]:
  w = count_allwords(i)
  c = word_length(i)
  s = count_sentences(i)
  term1 = 0.0588*c*25/w 
  term3 = 0.296*s*25/w 
  cl_score = term1-15.8-term3
  CL.append(cl_score)  
```

```python
len(CL)
```

```python
def perceptron_classifier (X_train_std, y_train, X_test_std, y_test):
  p = Perceptron(random_state=150)
  p.fit(X_train_std, y_train)
  predictions_train = p.predict(X_train_std)
  predictions_test = p.predict(X_test_std)
  train_score = accuracy_score(predictions_train, y_train)
  #print("score on train data: ", train_score)
  test_score = f1_score(predictions_test, y_test, average='weighted' )
  print("Perceptron F1 score:", test_score)  
```

```python
def classifcation_report_train_test(y_train, y_train_pred, y_test, y_test_pred):

    print('''
            =========================================
               CLASSIFICATION REPORT FOR TRAIN DATA
            =========================================
            ''')
    print(classification_report(y_train, y_train_pred, digits=4))

    print('''
            =========================================
               CLASSIFICATION REPORT FOR TEST DATA
            =========================================
            ''')
    print(classification_report(y_test, y_test_pred, digits=4))
```

```python
scores = pd.DataFrame(columns=['Model', 'Train_Accuracy', 'Train_Recall', 'Train_Precision', 'Train_F1_Score', 
                               'Test_Accuracy', 'Test_Recall', 'Test_Precision', 'Test_F1_Score'])

def get_metrics(train_actual, train_predicted, test_actual, test_predicted, model_description, dataframe):

    train_accuracy  = accuracy_score(train_actual, train_predicted)
    train_recall    = recall_score(train_actual, train_predicted, average="weighted")
    train_precision = precision_score(train_actual, train_predicted, average="weighted")
    train_f1score   = f1_score(train_actual, train_predicted, average="weighted")
    
    test_accuracy   = accuracy_score(test_actual, test_predicted)
    test_recall     = recall_score(test_actual, test_predicted, average="weighted")
    test_precision  = precision_score(test_actual, test_predicted, average="weighted")
    test_f1score    = f1_score(test_actual, test_predicted, average="weighted")

    dataframe       = dataframe.append(pd.Series([model_description, 
                                                  train_accuracy, train_recall, train_precision, train_f1score,
                                                  test_accuracy, test_recall, test_precision, test_f1score],
                                                 index=scores.columns ), 
                                       ignore_index=True)

    return(dataframe)
```
